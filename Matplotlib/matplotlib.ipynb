<h2>Intro to pyplot</h2>
<p><a class="reference internal" title="matplotlib.pyplot" href="https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot"><code class="xref py py-mod docutils literal notranslate"><span class="pre">matplotlib.pyplot</span></code></a>&nbsp;is a collection of functions that make matplotlib work like MATLAB. Each&nbsp;<code class="docutils literal notranslate"><span class="pre">pyplot</span></code>&nbsp;function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.</p>
<p>In&nbsp;<a class="reference internal" title="matplotlib.pyplot" href="https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot"><code class="xref py py-mod docutils literal notranslate"><span class="pre">matplotlib.pyplot</span></code></a>&nbsp;various states are preserved across function calls, so that it keeps track of things like the current figure and plotting area, and the plotting functions are directed to the current axes (please note that "axes" here and in most places in the documentation refers to the&nbsp;<em>axes</em>&nbsp;<a class="reference internal" href="https://matplotlib.org/3.3.1/tutorials/introductory/usage.html#figure-parts"><span class="std std-ref">part of a figure</span></a>&nbsp;and not the strict mathematical term for more than one axis).</p>

Pyplot Tutorial - https://matplotlib.org/3.3.1/tutorials/introductory/pyplot.html
